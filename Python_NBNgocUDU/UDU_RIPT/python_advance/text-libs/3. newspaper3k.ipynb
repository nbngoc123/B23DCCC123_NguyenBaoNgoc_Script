```python
# Crawl data
from newspaper import Article

url = 'https://thanhnien.vn/tai-chinh-kinh-doanh/gia-vang-hom-nay-2692021-chuyen-gia-du-bao-se-tiep-tuc-giam-1454975.html'
article = Article(url)

article.download()
print(article.html)
```

```python
article.parse()
print(article.text)
```

```python
content = article.text.replace("\n", "")
print(content)
```

```python
import nltk
nltk.download('punkt')
```
