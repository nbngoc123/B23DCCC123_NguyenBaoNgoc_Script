```python
class AdditionValue:        
    def __init__(self, value, maxValue=0):
        self.value = value
        self.maxValue = maxValue

    def __iter__(self):
        self.n = 0
        return self

    def __next__(self):
        if self.n <= self.maxValue:
            result = self.n + self.value
            self.n += 1
            
            return result
        else:
            raise StopIteration
```

```python
# create an object
numbers = AdditionValue(5, 2)

for number in numbers:
    print(number)
```

```python
# create an object
numbers = AdditionValue(5, 2)

# create an iterator
a_iter = iter(numbers)

# get iterator elements
print(next(a_iter))
print(next(a_iter))
print(next(a_iter))
```
